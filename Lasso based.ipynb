```
population_training_df = pd.read_csv(PATH+'population_training.csv', encoding='cp1252').dropna(axis=0)
population_testing_df = pd.read_csv(PATH+'population_sample.csv', encoding='cp1252').dropna(axis=0)
PopTestFULL_df = pd.read_csv(PATH+'population_testing.csv', encoding='cp1252').dropna(axis=0)
coef_df = pd.read_csv(PATH+'population_parameters_Lasso.csv', encoding='cp1252').dropna(axis=0)
```

```
train_data = population_training_df[:].drop('Id', axis = 1)


indz = 0
#going through each country.
for country in (population_training_df.drop(columns='Id')):
  
  X_train = population_training_df.drop(columns=[country,'Id'], axis = 1)
  y_train = population_training_df[country]

  #setting up the lasso to allow for setting changes.
  lasso = Lasso(fit_intercept=True)
  lasso.fit(X_train, y_train)
  train_score = lasso.score(X_train, y_train)
  coeff_used = np.sum(lasso.coef_!=0)
  found4 = False
  found2 = False

  # creating a list of alphas thats 150 elements long, this will help us select the correct alpha to use the proper number of features.
  alphas = 55**np.linspace(10,5,150)*0.05
  
  print('\nCountry tested: ',country)
  for alpha in alphas:
    #i'm prioritizing 4 country features, and then 3, and then anything between 1 and 2 below.
    if(found4 == False):
      lasso001 = Lasso(alpha, max_iter=10e7)
      lasso001.fit(X_train,y_train)
      train_score001=lasso001.score(X_train,y_train)
      coeff_used001 = np.sum(lasso001.coef_!=0)
    if(coeff_used001 == 4 and found4 == False):
      print('Alpha used: ',alpha)
      print("number of features used: for alpha above:", coeff_used001)
      found4 = True

  if found4 == False:
    found3 = False
    for alpha in alphas:
      if(found3 == False):
        lasso001 = Lasso(alpha, max_iter=10e5)
        lasso001.fit(X_train,y_train)
        train_score001=lasso001.score(X_train,y_train)
        #test_score001=lasso001.score(X_test,y_test)
        coeff_used001 = np.sum(lasso001.coef_!=0)
      if(coeff_used001 == 3 and found3 == False):
        print('Alpha used: ',alpha)
        print("number of features used: for alpha above:", coeff_used001)
        found3 = True

    if found3 == False:
      for alpha in alphas:
        if(found2 == False):
          lasso001 = Lasso(alpha, max_iter=10e5)
          lasso001.fit(X_train,y_train)
          train_score001=lasso001.score(X_train,y_train)
          coeff_used001 = np.sum(lasso001.coef_!=0)
        if(coeff_used001 <= 2 and coeff_used001 >=1 and found2 == False):
          print('Alpha used: ',alpha)
          print("number of features used: for alpha above:", coeff_used001)
          found2 = True
  #creating the full prediction.
  pred = lasso001.predict(PopTestFULL_df.drop(columns=[country,'Id'],axis=1))
  population_testing_df[country] = pred

  #creating the coefficient list and using the INDZ variable to declare where you need to insert a 0 being the country itself in the listing.
  coeflist = list(lasso001.coef_)
  coeflist.insert(indz,0)

  coef_df[country] = (coeflist)  
  indz = indz + 1

```
